```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from datetime import datetime

def scroll_and_click(driver, by, value):
    elem = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((by, value)))
    driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", elem)
    time.sleep(0.8)
    elem.click()

def scroll_and_fill(driver, by, value, text):
    elem = WebDriverWait(driver, 20).until(EC.presence_of_element_located((by, value)))
    driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", elem)
    time.sleep(0.5)
    elem.clear()
    elem.send_keys(text)
    time.sleep(0.5)

def scroll_and_click_by_text(driver, text):
    xpath = f"//*[contains(text(), '{text}')]"
    elem = WebDriverWait(driver, 20).until(
        EC.visibility_of_element_located((By.XPATH, xpath))
    )
    driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", elem)
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, xpath)))
    time.sleep(0.8)
    elem.click()


USERNAME = "salvador.navas@unican.es"
PASSWORD = "*****"

for year in range(1979, datetime.now().year + 1):
    # Iniciar navegador
    driver = webdriver.Chrome()
    driver.get("https://ewds.climate.copernicus.eu/datasets/cems-glofas-historical?tab=download")
    driver.maximize_window()

    # Paso 1: Aceptar cookies si aparece
    try:
        scroll_and_click(driver, By.CLASS_NAME, "bzjMTu")  # Botón "Accept all"
    except:
        pass

    # Paso 2: Login si es necesario
    try:
        scroll_and_click(driver, By.XPATH, "//button[contains(text(), 'Login/register')]")
        WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.ID, "username")))

        driver.find_element(By.ID, "username").send_keys(USERNAME)
        driver.find_element(By.ID, "password").send_keys(PASSWORD)
        driver.find_element(By.ID, "kc-login").click()
        time.sleep(5)  # Esperar a que completes login manual o automatizar si lo deseas
    except:
        pass

    # Paso 3: Seleccionar versión 4.0
    scroll_and_click(driver, By.ID, "Operational_version_4_0")

    # Paso 4: Seleccionar modelo LISFLOOD
    scroll_and_click(driver, By.ID, "hydrological_model_lisflood")

    # Paso 5: Seleccionar "Consolidated"
    scroll_and_click(driver, By.ID, "product_type_consolidated")

    # Paso 7: Descargar año a año
    year_id = f"hyear_{year}"
    scroll_and_click(driver, By.ID, year_id)

    # Seleccionar todos los meses y días
    scroll_and_click(driver, By.XPATH, "//button[@aria-label='Select all Month']")
    scroll_and_click(driver, By.XPATH, "//button[@aria-label='Select all Day']")

    scroll_and_click(driver, By.ID, "area")

    # Paso 6: Definir área
    coords = {
        "area_n": "15",
        "area_w": "-80",
        "area_e": "-65",
        "area_s": "-5"
    }
    for field_id, value in coords.items():
        scroll_and_fill(driver, By.ID, field_id, value)

    scroll_and_click(driver, By.ID, "netcdf")

    # Paso 8: Descargar
    scroll_and_click(driver, By.XPATH, "//button[contains(text(), 'Submit form')]")
    print("Descarga iniciada.")

    # Espera para descarga
    time.sleep(10)

      # Cerrar navegador
    driver.quit()
```
